* <b>Role</b>: contains all the user-item tuples corresponding to their ratings, that are used for the training phase.

### ml-1m.test.rating:
* <b>Role</b>: contains all the user-item tuples corresponding to their ratings (positive instances), that are used for the test phase.
* <b>Structure</b>: each line is a training instance comprised of userID, itemID, rating (1-5), and timestamp (if included).
* <b>How it was created</b>: this file contains only one rating for each user.

* <b>Role</b>: contains the user-item tuples corresponding to their ratings (negative instances), that are used for the test phase.

```python
# Helper Functions
def load_rating_file_as_list(filename):
    ratingList = []
    
    with open(filename, 'r') as f:
        line = f.readline()
        while line != None and line != "":
            arr = line.split('\t')
            user, item = int(arr[0]), int(arr[1])
            ratingList.append([user, item])
            line = f.readline()
            
    return ratingList
    
def load_negative_file(filename):
    negativeList = []
    
    with open(filename, 'r') as f:
        line = f.readline()
        while line != None and line != "":
            arr = line.split('\t')
            negatives = []
            for x in arr[1: ]:
                negatives.append(int(x))
            negativeList.append(negatives)
            line = f.readline()
            
    return negativeList

def load_rating_file_as_matrix(filename):
    num_users, num_items = 0, 0
    
    with open(filename, 'r') as f:
        line = f.readline()
        while line != None and line != "":
            arr = line.split('\t')
            u, i = int(arr[0]), int(arr[1])
            num_users = max(num_users, u)
            num_items = max(num_items, i)
            line = f.readline()

    mat = sp.dok_matrix((num_users + 1, num_items + 1), dtype=np.float32)
    with open(filename, 'r') as f:
        line = f.readline()
        while line != None and line != "":
            arr = line.split('\t')
            user, item, rating = int(arr[0]), int(arr[1]), float(arr[2])
            if (rating > 0):
                mat[user, item] = 1.0
            line = f.readline()  

    return mat

def get_train_instances(train, num_negatives):
    user_input, item_input, labels = [],[],[]
    num_users = train.shape[0]
    
    for (u, i) in train.keys():
        # positive instance
        user_input.append(u)
        item_input.append(i)
        labels.append(1)
        
        # negative instances
        for t in range(num_negatives):
            j = np.random.randint(num_items)
            while (u, j) in train:
                j = np.random.randint(num_items)
            user_input.append(u)
            item_input.append(j)
            labels.append(0)

    return np.array(user_input), np.array(item_input), np.array(labels)
```

<br><br><br><br>
**Matrix Factorization (MF)**

<br>



```python
def get_MF_model(num_users, num_items, latent_dim=20):
    # input layers - users & items
    user_input = Input(shape=(1,), dtype='int32', name='user_input')
    item_input = Input(shape=(1,), dtype='int32', name='item_input')
    
    # embedding users' & items' layers
    embedding_user = Embedding(input_dim=num_users, output_dim=latent_dim, name='user_embedding', input_length=1)
    embedding_item = Embedding(input_dim=num_items, output_dim=latent_dim, name='item_embedding', input_length=1)
    
    # flattening the embedding users' & items'
    user_latent = Flatten()(embedding_user(user_input))
    item_latent = Flatten()(embedding_item(item_input))
    
    # using Dot (instead of multiply) on the flattened vectors
    user_item_multiply = keras.layers.dot([user_latent,item_latent], axes=1, normalize=False)
    
    # create the model
    model = Model(inputs=[user_input, item_input], outputs=user_item_multiply, name='MF_model')
    
    return model
```

```python
model_MF = get_MF_model(num_users, num_items)
model_MF.summary()
```

<br><br><br><br><br><br>
**Multi Layer Perceptron (MLP)**
<br>


```python
def get_MLP_model(num_users, num_items, latent_dim=8, regs=[0,0,0], layers=[32,16,8]):
    # input layers - users & items
    user_input = Input(shape=(1,), dtype='int32', name='user_input')
    item_input = Input(shape=(1,), dtype='int32', name='item_input')
    
    # embedding users' & items' layers
    embedding_user = Embedding(input_dim=num_users, output_dim=latent_dim, name='user_embedding',
                                   embeddings_regularizer = l2(regs[0]), input_length=1, 
                                   embeddings_initializer=RandomNormal(mean=0.0, stddev=0.01))
    embedding_item = Embedding(input_dim = num_items, output_dim = latent_dim, name='item_embedding',
                                   embeddings_regularizer = l2(regs[0]), input_length=1,
                                   embeddings_initializer=RandomNormal(mean=0.0, stddev=0.01))
    
    # flattening the embedding users' & items' vectors
    user_latent = Flatten()(embedding_user(user_input))
    item_latent = Flatten()(embedding_item(item_input))    
    
    # concatenation the flattened layers
    concat_vectors = Concatenate(axis=-1)([user_latent, item_latent])
    
    # initial MLP layers
    for idx in range(len(layers)):
        layer = Dense(layers[idx], kernel_regularizer = l2(regs[idx]), activation='relu', name=f'layer{idx}')
        concat_vectors = layer(concat_vectors)
    
    # output prediction dense layer
    prediction = Dense(1, activation='sigmoid', kernel_initializer='lecun_uniform', name='prediction')(concat_vectors)
    
    # create the model
    model = Model(inputs=[user_input, item_input], outputs=prediction, name='MLP_model')
    
    return model
```

```python
model_MLP = get_MLP_model(num_users, num_items)
model_MLP.summary()
```

**Generalized Matrix Factorization (GMF)**
<br>


```python
def get_GMF_model(num_users, num_items, latent_dim=20, regs=[0,0], dropout=0.4):
    # input layers - users & items
    user_input = Input(shape=(1,), dtype='int32', name='user_input')
    item_input = Input(shape=(1,), dtype='int32', name='item_input')

    # embedding users' & items' layers
    embedding_user = Embedding(input_dim=num_users, output_dim=latent_dim, name='user_embedding',
                                  embeddings_regularizer=l2(regs[0]), input_length=1,
                                  embeddings_initializer=RandomNormal(mean=0.0, stddev=0.01))
    embedding_item = Embedding(input_dim=num_items, output_dim=latent_dim, name='item_embedding',
                                  embeddings_regularizer=l2(regs[1]), input_length=1,
                                  embeddings_initializer=RandomNormal(mean=0.0, stddev=0.01))
    
    # flattening the embedding users' & items' vectors
    user_latent = Flatten()(embedding_user(user_input))
    item_latent = Flatten()(embedding_item(item_input))
    
    # multiply the flattened vectors
    multiply_layer = Multiply()([user_latent, item_latent]) 
    
    # dropout layer
    dropout_layer = Dropout(dropout)(multiply_layer)
    
    # output prediction dense layer
    prediction = Dense(1, activation='sigmoid', kernel_initializer='lecun_uniform', name='prediction')(dropout_layer)
    
    # create the model
    model = Model(inputs=[user_input, item_input], outputs=prediction, name='GMF_model')
    
    return model
```

```python
model_GMF = get_GMF_model(num_users, num_items)
model_GMF.summary()
```

**NeuroMatrixFactorization (NMF)**

<br>



```python
def get_NMF_model(num_users, num_items, latent_dim_GMF=8, latent_dim_MLP=8, reg_GMF=[0,0], regs_MLP=[0,0,0], layers=[32,16,8]):
    # input layers - users & items
    user_input = Input(shape=(1,), dtype='int32', name='user_input')
    item_input = Input(shape=(1,), dtype='int32', name='item_input')
    
    # embedding users' & items' layers
    mf_embedding_user = Embedding(input_dim=num_users, output_dim=latent_dim_GMF, name = 'MF_user_embedding',
                                  embeddings_regularizer = l2(reg_GMF[0]), input_length=1,
                                  embeddings_initializer=RandomNormal(mean=0.0, stddev=0.01))
    mf_embedding_item = Embedding(input_dim=num_items, output_dim=latent_dim_GMF, name='MF_item_embedding',
                                  embeddings_regularizer = l2(reg_GMF[1]), input_length=1,
                                  embeddings_initializer=RandomNormal(mean=0.0, stddev=0.01))
    mlp_embedding_eser = Embedding(input_dim=num_users, output_dim=latent_dim_MLP, name='MLP_user_embedding',
                                   embeddings_regularizer = l2(regs_MLP[0]), input_length=1,
                                   embeddings_initializer=RandomNormal(mean=0.0, stddev=0.01))
    mlp_embedding_item = Embedding(input_dim=num_items, output_dim=latent_dim_MLP, name='MLP_item_embedding',
                                   embeddings_regularizer = l2(regs_MLP[0]), input_length=1,
                                   embeddings_initializer=RandomNormal(mean=0.0, stddev=0.01))
    
    # MF flattened layers
    mf_user_latent = Flatten()(mf_embedding_user(user_input))
    mf_item_latent = Flatten()(mf_embedding_item(item_input))
    mf_vector = Multiply()([mf_user_latent, mf_item_latent])
    
    # MLP flattened layers 
    mlp_user_latent = Flatten()(mlp_embedding_eser(user_input))
    mlp_item_latent = Flatten()(mlp_embedding_item(item_input))
    mlp_vector = Concatenate(axis=-1)([mlp_user_latent, mlp_item_latent])
    
    # initial MLP dense layers
    for idx in range(len(layers)):
        layer =  Dense(layers[idx], kernel_regularizer=l2(regs_MLP[idx]), activation='tanh', name=f'layer{idx}')
        mlp_vector = layer(mlp_vector)
        
    # concatenation all the layers
    concat_layer = Concatenate(axis=-1)([mf_vector, mlp_vector])
    
    # output prediction dense layer
    prediction = Dense(1, activation='sigmoid', kernel_initializer='lecun_uniform', name="prediction")(concat_layer)    
    
    # create the model
    model = Model(inputs=[user_input, item_input], outputs=prediction, name='NMF_model')
    
    return model
```

```python
model_NMF = get_NMF_model(num_users, num_items)
model_NMF.summary()
```

## b. Train and evaluate the recommendations accuracy of the models:

Compare the learning curve and recommendations accuracy using NDCG and MRR metrics with cutoff values of 5 and 10.   
Discuss the comparison. 

**Metrics:**
- HitRatio
- nDCG
- MRR

**Evaluation:**

```python
# Helper Functions
class Evaluation(object):
    
    def __init__(self, model, test_ratings, test_negatives, top_n):
        self.model = model
        self.test_ratings = test_ratings
        self.test_negatives = test_negatives
        self.top_n = top_n

    def evaluate_model(self):
        hits10, hits5, ndcgs10, ndcgs5, mrrs10, mrrs5 = [], [], [], [], [], []
        
        # evalute ratings
        for idx in range(len(self.test_ratings)):
            hit10, hit5, ndcg10, ndcg5, mrr10, mrr5 = self._eval_one_rating(idx)
            
            hits10.append(hit10)
            hits5.append(hit5)
            
            ndcgs10.append(ndcg10)
            ndcgs5.append(ndcg5)
            
            mrrs10.append(mrr10)
            mrrs5.append(mrr5)
        
        # compute hit ratio
        hit10 = np.mean(hits10)
        hit5 = np.mean(hits5)
        
        # compute ndcg
        ndcg10 = np.mean(ndcgs10)
        ndcg5 = np.mean(ndcg5)
        
        # compute mrr
        mrr10 = np.mean(mrrs10)       
        mrr5 = np.mean(mrrs5)
        
        return hit10, hit5, ndcg10, ndcg5, mrr10, mrr5

    def _eval_one_rating(self, idx):
        rating = self.test_ratings[idx]
        items = self.test_negatives[idx]
        u = rating[0]
        gtItem = rating[1]
        items.append(gtItem)

        # compute the prediction scores
        map_item_score = {}
        users = np.full(len(items), u, dtype='int32')
        predictions = self.model.predict([users, np.array(items)], batch_size=100, verbose=0)
        
        for i in range(len(items)):
            item = items[i]
            map_item_score[item] = list(predictions[i])
        
        # evaluate top rank list - cutoff values of 10
        ranklist = heapq.nlargest(self.top_n, map_item_score, key=map_item_score.get)
        hit10 = Evaluation._getHitRatio(ranklist, gtItem)        
        ndcg10 = Evaluation._getNDCG(ranklist, gtItem)
        mrr10 = Evaluation._getMrr(ranklist, gtItem)
        
        # evaluate top rank list - cutoff values of 5
        ranklist = heapq.nlargest(self.top_n//2, map_item_score, key=map_item_score.get)
        hit5 = Evaluation._getHitRatio(ranklist, gtItem)  
        ndcg5 = Evaluation._getNDCG(ranklist, gtItem)
        mrr5 = Evaluation._getMrr(ranklist, gtItem)
        
        return hit10, hit5, ndcg10, ndcg5, mrr10, mrr5

    @staticmethod
    def _getHitRatio(ranklist, gtItem):
        for item in ranklist:
            if item == gtItem:
                return 1
        return 0
    
    @staticmethod
    def _getNDCG(ranklist, gtItem):
        for i in range(len(ranklist)):
            item = ranklist[i]
            if item == gtItem:
                return np.log2(2) / np.log2(i + 2)
        return 0
    
    @staticmethod
    def _getMrr(ranklist, gtItem):
        for i in range(len(ranklist)):
            item = ranklist[i]
            if item == gtItem:
                return 1 / (i + 1)
        return 0
```

```python
def train_evaluate_model(model, model_evaluation, epochs, batch_size, verbose):
    # present the current model
    print(f'--- model: {model.name} ---')
    
    # store fit time
    start = time()    

    # fit model
    early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    model_fit_res = model.fit([X_user_train, X_item_train], y_train, \
                              validation_data = ([X_user_test, X_item_test], y_test), \
                              batch_size=batch_size, \
                              epochs=epochs, \
                              callbacks=[early_stopping], verbose=1, shuffle=True)
        
#     model_fit_res = model.fit([user_train, item_train], labels_train, validation_split=0.3, \
#                               batch_size=batch_size, epochs=epochs, \
#                               callbacks=[early_stopping], verbose=1)

    # store train & validation losses 
    losses = model_fit_res.history['loss']
    val_losses = model_fit_res.history['val_loss']

    # evaluate model
    model_hit10, model_hit5, model_ndcg10, model_ndcg5, model_mrr10, model_mrr5 = \
        model_evaluation.evaluate_model()
    
    train_eval_time = time() - start
    
    # present model's evaluation results
    print(f'\nModel: {model.name}: Train & Evaluate is Done. \nEvaluation Results: HIT_RATIO10: {model_hit10:.4f}, HIT_RATIO5: {model_hit5:.4f}, NDCG10: {model_ndcg10:.4f}, NDCG5: {model_ndcg5:.4f}, MRR10: {model_mrr10:.4f}, MRR5: {model_mrr5:.4f}')
    
    return model_hit10, model_hit5, model_ndcg10, model_ndcg5, model_mrr10, model_mrr5, \
            losses, val_losses, train_eval_time
    
```

The MRR and NDCG metrics are used for evaluating the ordered output of a given model. The order of the output recommendations is by the most relevant to the least relevant item.

According to this assignment and the results above, we can observe that when using the implicit ranking approach, the order of the relevancy was ignored since every item that each user has ranked we consider as relevant, and every item that wasn't ranked we consider as irrelevant.

Therefore, using the MRR and NDCG metrics for evaluating measures in the implicit case will be less accurate since the basic assumption that the order of the items is relevant is ignored.

#### a. One of the enhancements presented in the Neural Collaborative Filtering paper is the usage of probabilistic activation function (the sigmoid) and binary cross entropy loss function.   

Select one of the models you implemented in question 2 and change the loss function to a `Mean Squared Error` and the activation function of the last layer to `RELU`.   

Train the model and evaluate it in a similar way to what you did in question 2. 
Compare the results and discuss.

```python
def get_updated_NMF_model(num_users, num_items, latent_dim_GMF=8, latent_dim_MLP=8, reg_GMF=[0,0], regs_MLP=[0,0,0], layers=[32,16,8]):
    # input layers - users & items
    user_input = Input(shape=(1,), dtype='int32', name='user_input')
    item_input = Input(shape=(1,), dtype='int32', name='item_input')
    
    # embedding users' & items' layers
    mf_embedding_user = Embedding(input_dim=num_users, output_dim=latent_dim_GMF, name = 'MF_user_embedding',
                                  embeddings_regularizer = l2(reg_GMF[0]), input_length=1,
                                  embeddings_initializer=RandomNormal(mean=0.0, stddev=0.01))
    mf_embedding_item = Embedding(input_dim=num_items, output_dim=latent_dim_GMF, name='MF_item_embedding',
                                  embeddings_regularizer = l2(reg_GMF[1]), input_length=1,
                                  embeddings_initializer=RandomNormal(mean=0.0, stddev=0.01))
    mlp_embedding_eser = Embedding(input_dim=num_users, output_dim=latent_dim_MLP, name='MLP_user_embedding',
                                   embeddings_regularizer = l2(regs_MLP[0]), input_length=1,
                                   embeddings_initializer=RandomNormal(mean=0.0, stddev=0.01))
    mlp_embedding_item = Embedding(input_dim=num_items, output_dim=latent_dim_MLP, name='MLP_item_embedding',
                                   embeddings_regularizer = l2(regs_MLP[0]), input_length=1,
                                   embeddings_initializer=RandomNormal(mean=0.0, stddev=0.01))
    
    # MF flattened layers
    mf_user_latent = Flatten()(mf_embedding_user(user_input))
    mf_item_latent = Flatten()(mf_embedding_item(item_input))
    mf_vector = Multiply()([mf_user_latent, mf_item_latent])
    
    # MLP flattened layers 
    mlp_user_latent = Flatten()(mlp_embedding_eser(user_input))
    mlp_item_latent = Flatten()(mlp_embedding_item(item_input))
    mlp_vector = Concatenate(axis=-1)([mlp_user_latent, mlp_item_latent])
    
    # initial MLP dense layers
    for idx in range(len(layers)):
        layer =  Dense(layers[idx], kernel_regularizer=l2(regs_MLP[idx]), activation='tanh', name=f'layer{idx}')
        mlp_vector = layer(mlp_vector)
        
    # concatenation all the layers
    concat_layer = Concatenate(axis=-1)([mf_vector, mlp_vector])
    
    # output prediction dense layer - contains the 'relu' activation function
    prediction = Dense(1, activation='relu', kernel_initializer='lecun_uniform', name="prediction")(concat_layer)    
    
    # create the model
    model = Model(inputs=[user_input, item_input], outputs=prediction, name='updated_NMF_model')
    
    return model
```

```python
# The Updated NMF Model - contains the relu activation the the mse loss functions
model_updated_NMF = get_updated_NMF_model(num_users, num_items, latent_dim_GMF=num_factors, latent_dim_MLP=num_factors)
model_updated_NMF.compile(optimizer=Adam(lr=learning_rate), loss='mse')
model_updated_NMF.summary()
```

### Conclusions:

In this assignment, we were required to implement, perform and evaluate Deep Learning components for collaborative filtering & hybrid approach recommenders. We followed the given dataset to use an implicit interaction of the items (for each user).

We build multiple recommenders DL models following each architecture separately:
* Matrix Factorization (MF).
* Multi-layer perceptron (MLP).
* Generalized Matrix Factorization (GMF).
* NeuroMatrixFactorization (NMF).

According to the overall results above, 
* MF achieved poor results compared to the other models.
* NMF achieved the best scores among the models and its time performances were took more additional time followed by the complexity of its architecture.
* According to the Loss Functions' plots, we can observe that there are some plots with slightly overfitting, hence, we used the "EarlyStopping" callback which saves us the best wights for our model.
* The updated NMF model, using the "relu" activation and the "mse" loss function, achieved almost the same results compared to the base NMF.

#### NMF vs. Updated NMF:
The differences between the losses were caused by the calculation formula. The point is, using implicit data (consider as a classification problem) the MSE loss function did not penalize enough for misclassifications (and the Updated NMF model was slightly overfitting according to the Training and Validation plot).

The differences between the activation functions were caused by the technique each function performs. 
* The "sigmoid" function restricts the neurons and can saturate gradients when the firing rate is either close to 0 or 1.
* The "relu" function can greatly accelerate the convergence of stochastic gradient descent compared to the sigmoid, it is sparsity and less restricted and simple (also has better performance due to its simplicity).

#### Evaluation Results:

We used the Hit Ration, NDCG, and MRR Metrix for evaluating our DL models (using 5 and 10 cutoffs).
As already mentioned above, compared to our previous assignments, when performing these metrics on implicit datasets, we achieve less accurate results since the basic assumption that the (rank) order of the items is relevant was ignored.
