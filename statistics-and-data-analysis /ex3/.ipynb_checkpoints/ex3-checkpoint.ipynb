### 1.F Answer

I think the situation described here is impossible since the two correlation methods (Spearman and Kendall) are based on the rank's order and not the exact values of the variables. They are not sensitive to changes in the measurements' values, and they assume that there is a monotonic relationship between the two variables, i.e. they "act"/"behave" almost the same base on the monotonic relationship.
Therefore, we cannot provide an example in which the Spearman correlation result will be 1.2 higher than the Kendall correlation, which means we will get a completely different correlation, for instance, the Spearman result will be 1 (the max positive correlation), while the Kendall result will be at least -0.2 (which can be no correlation or a lower negative correlation).

```python
# get the uci dataset
df_uci = pd.read_csv('heart.csv', sep='\t')

# numerical features
num_ft = np.array(['age', 'trestbps', 'chol', 'thalach', 'oldpeak'])

# categorical features
cat_ft = np.array(['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])

# fill missing values
df_uci[num_ft] = df_uci[num_ft].fillna(df_uci[num_ft].mean())
df_uci[cat_ft] = df_uci[cat_ft].fillna(df_uci[cat_ft].mode().iloc[0])

print(df_uci.info())
```

```python
def calc_confidence_interval(df_uci_num, num_feature):
    df_healthy_feature = df_uci_num[df_uci_num['target'] == 0][num_feature]
    df_disease_feature = df_uci_num[df_uci_num['target'] == 1][num_feature]

    n_healthy_feature = df_healthy_feature.shape[0]
    n_disease_feature = df_disease_feature.shape[0]
    p_rohat_healthy_feature = df_healthy_feature.mean()
    p_rohat_disease_feature = df_disease_feature.mean()

    print(f'The mean {num_feature} of the healthy population empirical: {p_rohat_healthy_feature:.4f}')
    print(f'The mean {num_feature} of the disease population empirical: {p_rohat_disease_feature:.4f}')

    alpha = 0.05
    gamma = stats.norm.ppf(q=1 - alpha / 2)

    sigma_rohat_healthy_feature = df_healthy_feature.std(ddof=1)
    error_margin_healthy_feature = gamma * sigma_rohat_healthy_feature
    left_healthy_feature = p_rohat_healthy_feature - error_margin_healthy_feature
    right_healthy_feature = p_rohat_healthy_feature + error_margin_healthy_feature
    healthy_feature_interval = (left_healthy_feature, right_healthy_feature)
    print(f'The mean {num_feature} of the health population is in the interval: {healthy_feature_interval}')

    sigma_rohat_disease_feature = df_disease_feature.std(ddof=1)
    error_margin_disease_feature = gamma * sigma_rohat_disease_feature
    left_disease_feature = p_rohat_disease_feature - error_margin_disease_feature
    right_disease_feature = p_rohat_disease_feature + error_margin_disease_feature
    disease_feature_interval = (left_disease_feature, right_disease_feature)
    print(f'The mean {num_feature} of the disease population is in the interval: {disease_feature_interval}')

    empirical_delta = p_rohat_healthy_feature - p_rohat_disease_feature
    error_margin_delta = gamma * np.sqrt(((sigma_rohat_healthy_feature ** 2) / n_healthy_feature) +
                                         ((sigma_rohat_disease_feature ** 2) / n_disease_feature))

    left_delta = empirical_delta - error_margin_delta
    right_delta = empirical_delta + error_margin_delta
    delta_interval = (left_delta, right_delta)
    print(f'The real delta of the mean {num_feature} health and the mean {num_feature} disease populations is in the '
          f'interval:')
    print(delta_interval)


# computing the population means with 95% confidence for the numerical features
num_ft_target = np.array(['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'target'])
df_uci_num = df_uci[num_ft_target]

for num_feature in num_ft_target:
    if num_feature != 'target':
        print(f'------- {num_feature} -------')
        calc_confidence_interval(df_uci_num, num_feature)
```

```python
import warnings
warnings.filterwarnings('ignore')

def best_fit_distribution(data, bins=200, ax=None):
    y, x = np.histogram(data, bins=bins, density=True)
    x = (x + np.roll(x, -1))[:-1] / 2.0

    # list of all the continuous distributions
    distributions = [        
        stats.alpha,stats.anglit,stats.arcsine,stats.beta,stats.betaprime,stats.bradford,stats.burr,stats.cauchy,stats.chi,stats.chi2,stats.cosine,
        stats.dgamma,stats.dweibull,stats.erlang,stats.expon,stats.exponnorm,stats.exponweib,stats.exponpow,stats.f,stats.fatiguelife,stats.fisk,
        stats.foldcauchy,stats.foldnorm,stats.frechet_r,stats.frechet_l,stats.genlogistic,stats.genpareto,stats.gennorm,stats.genexpon,
        stats.genextreme,stats.gausshyper,stats.gamma,stats.gengamma,stats.genhalflogistic,stats.gilbrat,stats.gompertz,stats.gumbel_r,
        stats.gumbel_l,stats.halfcauchy,stats.halflogistic,stats.halfnorm,stats.halfgennorm,stats.hypsecant,stats.invgamma,stats.invgauss,
        stats.invweibull,stats.johnsonsb,stats.johnsonsu,stats.ksone,stats.kstwobign,stats.laplace,stats.levy,stats.levy_l,stats.levy_stable,
        stats.logistic,stats.loggamma,stats.loglaplace,stats.lognorm,stats.lomax,stats.maxwell,stats.mielke,stats.nakagami,stats.ncx2,stats.ncf,
        stats.nct,stats.norm,stats.pareto,stats.pearson3,stats.powerlaw,stats.powerlognorm,stats.powernorm,stats.rdist,stats.reciprocal,
        stats.rayleigh,stats.rice,stats.recipinvgauss,stats.semicircular,stats.t,stats.triang,stats.truncexpon,stats.truncnorm,stats.tukeylambda,
        stats.uniform,stats.vonmises,stats.vonmises_line,stats.wald,stats.weibull_min,stats.weibull_max,stats.wrapcauchy
    ]

    # best holders
    best_distribution = stats.norm
    best_params = (0.0, 1.0)
    best_sse = np.inf

    # estimate the distribution parameters from the data using the fit function (which using the MLE)
    for distribution in distributions:
        # return MLEs for shape (if applicable), location, and scale parameters from data
        params = distribution.fit(data)

        # set the parameters
        arg = params[:-2]
        loc = params[-2]
        scale = params[-1]

        # calculate the fitted PDF and the error
        pdf = distribution.pdf(x, loc=loc, scale=scale, *arg)
        sse = np.sum(np.power(y - pdf, 2.0))

        try:
            if ax:
                ax.plot(pd.Series(pdf, x))
        except Exception:
            pass

        # check if the current distribution is better
        if best_sse > sse > 0:
            best_distribution = distribution
            best_params = params
            best_sse = sse

    return best_distribution.name, best_params


def make_pdf(dist, params, size=1000):
    arg = params[:-2]
    loc = params[-2]
    scale = params[-1]

    start = dist.ppf(0.01, *arg, loc=loc, scale=scale) if arg else dist.ppf(0.01, loc=loc, scale=scale)
    end = dist.ppf(0.99, *arg, loc=loc, scale=scale) if arg else dist.ppf(0.99, loc=loc, scale=scale)

    x = np.linspace(start, end, size)
    y = dist.pdf(x, loc=loc, scale=scale, *arg)
    pdf = pd.Series(y, x)

    return pdf


def inspect_dist(data, title):
    # plot for comparison
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(16.0, 6.0))
    ax1.hist(data, density=True, bins=50, alpha=0.5)

    # store the plot limits
    y_lim = ax1.get_ylim()

    # model the data by finding the best fit distribution to it
    best_fit_name, best_fit_params = best_fit_distribution(data, 200, ax1)
    best_dist = getattr(stats, best_fit_name)

    # update the plots
    ax1.set_ylim(y_lim)
    ax1.set_title(f'All Trials for the Fitted Distributions - Numerical Feature: {title}')

    # visualization
    pdf = make_pdf(best_dist, best_fit_params)
    ax2.plot(pdf, label='PDF')
    ax2.hist(data, bins=50, density=True, alpha=0.5, label='Data')
    ax2.legend()

    param_names = (best_dist.shapes + ', loc, scale').split(', ') if best_dist.shapes else ['loc', 'scale']
    param_str = ', '.join(['{}={:0.2f}'.format(k, v) for k, v in zip(param_names, best_fit_params)])
    dist_str = '{}({})'.format(best_fit_name, param_str)
    ax2.set_title(f'{title}\n The best fit distribution: {dist_str}')

    plt.show()

for feature in num_ft:
    inspect_dist(df_uci[feature], feature)
```

```python
def plot_2d_Gaussian_pdf(means, cov, title):
    n = 100
    x1 = np.linspace(means[0] - 3 * np.sqrt(cov[0][0]), means[0] + 3 * np.sqrt(cov[0][0]), n)
    x2 = np.linspace(means[1] - 3 * np.sqrt(cov[1][1]), means[1] + 3 * np.sqrt(cov[1][1]), n)
    x1_v, x2_v = np.meshgrid(x1, x2)
    Xgrid = np.vstack([x1_v.ravel(), x2_v.ravel()]).T
    Y = mn.pdf(Xgrid, means, cov)
    fig, ax = plt.subplots()
    ax.pcolorfast(x1, x2, Y.reshape(x1_v.shape), alpha=0.5, cmap='Blues')
    ax.contour(x1_v, x2_v, Y.reshape(x1_v.shape), levels=[0.0001, 0.0003, 0.0005, 0.0007], alpha=0.3, colors='b')
    ax.axis('equal')
    ax.grid(alpha=0.2)
    plt.title(title)
    plt.show()

    
# present the pearson correlation of the numerical features (the default correlation in pandas.corr() is pearson)
plt.rcParams['figure.figsize'] = (8.0, 6.0)
sns.heatmap(df_uci[num_ft].corr(), annot=True, vmin=-1, vmax=1)
plt.show()
```

Nevertheless, this is all based on the empirical dataset, thus I will use the confidence interval (alpha = 0.05) to determine if the male population mean is larger/smaller than the female population mean, based on the disease data (target = 1).

Therefore, the null hypothesis may state that the male probability is greater or equal than the female probability, and the alternative hypothesis will state that the male probability is smaller than the female probability.

Based on the results below, if the upper confidence limit of the 'delta_interval' is less than zero, then we conclude that the observed ratio 'p_rohat_male' is statistically significantly smaller than 'p_rohat_female' with significance confidence of 95%, i.e. we can reject the null hypothesis with significance confidence of 95%.

```python
num_ft_target_sex = np.concatenate([num_ft, ['target', 'sex']])
df_uci_num = df_uci[num_ft_target_sex]
df_uci_num = df_uci_num[df_uci_num['target'] == 1]
n = df_uci_num.shape[0]
p_rohat_male = df_uci_num[df_uci_num['sex'] == 0]['sex'].count() / n
p_rohat_female = df_uci_num[df_uci_num['sex'] == 1]['sex'].count() / n

print(f'The male population empirical mean: {p_rohat_male:.4f}')
print(f'The female population empirical mean: {p_rohat_female:.4f}')

alpha = 0.05
gamma = stats.norm.ppf(q=1-alpha/2)

sigma_rohat_male = np.sqrt(p_rohat_male*(1-p_rohat_male)) / np.sqrt(n)
error_margin_male = gamma * sigma_rohat_male
left_male = p_rohat_male - error_margin_male
right_male = p_rohat_male + error_margin_male
male_interval = (left_male, right_male)
print(f'The real probability for the male population is in the interval: {male_interval}')

sigma_rohat_female = np.sqrt(p_rohat_female*(1-p_rohat_female)) / np.sqrt(n)
error_margin_female = gamma * sigma_rohat_female
left_female = p_rohat_female - error_margin_female
right_female = p_rohat_female + error_margin_female
female_interval = (left_female, right_female)
print(f'The real probability for the female population is in the interval: {female_interval}')

empirical_delta = p_rohat_male - p_rohat_female
error_margin_delta = gamma * np.sqrt(((p_rohat_male*(1-p_rohat_male))/n) +
                                     ((p_rohat_female*(1-p_rohat_female))/n))

left_delta = empirical_delta - error_margin_delta
right_delta = empirical_delta + error_margin_delta
delta_interval = (left_delta, right_delta)
print(f'the real delta of the male and female populations is in the interval: {delta_interval}')
```

### Question 3 - Heavy Tailed distributions (10 points)

Recall the definition of Heavy Tailed distribtion from the lectures.

*A distribution is said to have a heavy right tail if its tail probabilities vanish slower than any exponential*
$$ \forall t>0, \lim_{x\to\infty} e^{tx}P(X>x)=\infty $$

#### 3.A
Does the standard normal distribution have a heavy right tail? prove your answer.

#### No, the standard normal distribution don't have a heavy right tail.


##### Definition
The distribution of a random variable X with distribution function F is said to have a heavy (right) tail if the moment generating function of $X$, is infinite for all $t > 0$.

$\int_{-\infty}^\infty e^{tx} dF(x) = \infty ,\forall t > 0$

##### Proof
* Standard normal distribution: X ~ N(0,1)

$\int_{-\infty}^\infty e^{tx} \cdot \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2} x^2} dx =$ \
$\int_{-\infty}^\infty \frac{1}{\sqrt{2\pi}} \cdot e^{-\frac{1}{2} x^2 + tx} dx =$ \
$\int_{-\infty}^\infty \frac{1}{\sqrt{2\pi}} \cdot e^{-\frac{1}{2} (-2tx + x^2)} dx =$ \
$\int_{-\infty}^\infty \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2} (x - t)^2 + \frac{1}{2} t^2}$ \
$\int_{-\infty}^\infty \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2} (x - t)^2} e^{\frac{1}{2} t^2}$ \
$e^{\frac{1}{2} t^2} \cdot [\int_{-\infty}^\infty \frac{1}{\sqrt{2\pi}} \cdot e^{-\frac{1}{2} (x-t)^2} dx]$
$= e^{\frac{1}{2} t^2}$

##### Calculations
 * Since the integral of the probability density function converges to 1, then: $\int_{-\infty}^\infty \frac{1}{\sqrt{2\pi}} \cdot e^{-\frac{1}{2} (x-t)^2} dx = 1$
 * $-\frac{1}{2} x^2 + tx = -\frac{1}{2} (-2tx + x^2) = -\frac{1}{2} (x - t)^2 + \frac{1}{2} t^2$
 
##### Summary
Since the result is $e^{\frac{1}{2} t^2}$ which is not converges to $\infty$, then the standard normal distribution don't have a heavy right tail.

#### 3.B
Does the standard log-normal distribution have a heavy right tail? prove your answer.

#### Yes, the standard log-normal distribution have a heavy right tail.

##### Definition
The distribution of a random variable X with distribution function F is said to have a heavy (right) tail if the moment generating function of $X$, is infinite for all $t > 0$.

$\int_{-\infty}^\infty e^{tx} dF(x) = \infty ,\forall t > 0$

##### Proof
* Standard log-normal distribution: $Y = e^x$, X ~ N(0,1)

$\int_{0}^\infty e^{tx} \cdot \frac{1}{x\sqrt{2\pi}} e^{-\frac{1}{2}(ln(x))^2} dx =$ 

Based on the L'Hopital's rule, we can say that for any k > 0, and for every $x$ large enough, 
there is a constant C, which follows that,

$e^{tx} \cdot e^{-\frac{1}{2}(ln(x))^2} \geq C, \forall x > k$

Therefore,

$\int_{k}^\infty e^{tx} \frac{1}{x\sqrt{2\pi}} e^{-\frac{1}{2}(ln(x))^2} dx$
$\geq C \cdot \int_{k}^\infty \frac{1}{x} dx = C \cdot ln \vert_{k}^{\infty} = \infty$

##### Summary
Since the moment generating function of the log-normal distribution is grater than a function which is converges to $\infty$ (based on the L'Hopital's rule), thus the log-normal distribution is converges to $\infty$ as well, i.e. the standard log-normal distribution have a heavy right tail.

### Question 4 - CLT for Markov chains (20 points)

In this exercise you will construct trajectories of Markovian dice rolling results in the following way.<br>
The first roll, $X_0$, is $U(1...6)$, i.e. $P(X_0=i)=\pi_0(i)=1/6, \forall i\in{\{1,..,6\}}$<br>
After i rolls are determined the $i+1^{st}, X_{i+1},$ is drawn according to the row that corresponds to the value of $X_i$ in the matrix $T$ below. <br>
In other words, $T$ is the transition matrix of a Markov chain and the initial distribution is uniform.

\begin{equation*}
T = 
\begin{pmatrix}
0.4 & 0.2 & 0.1 & 0 & 0.1 & 0.2 \\
0.2 & 0.4 & 0.2 & 0.1 & 0 & 0.1 \\
0.1 & 0.2 & 0.4 & 0.2 & 0.1 & 0 \\
0 & 0.1 & 0.2 & 0.4 & 0.2 & 0.1 \\
0.1 & 0 & 0.1 & 0.2 & 0.4 & 0.2 \\
0.2 & 0.1 & 0 & 0.1 & 0.2 & 0.4
\end{pmatrix}
\end{equation*}

#### 4.A
Construct 1000 trajectories, each of length 30.
1. What do you expect the average value of all 30 numbers in a trajectory to be?
2. Compute the average value of each such trajectory. Draw a histogram of the 1000 numbers you received, using 20 bins.
3. What does the distribution look like? What are the empirical mean and the std?

##### Answer
1. Since the mean of the uniform distribution is: $E(x) = \frac{n+1}{2}$ 
I expect the average value of all 30 numbers in a trajectory to be 3.5. \
$E(X) = \frac{6+1}{2} = 3.5$ , when n = 6.
2. Please see the implementation below.
3. The distribution looks like a Normal distribution with the empirical mean of 3.5056 and the empirical standard deviation of 0.3135.

```python
def markov_chains(T, pi_0, states, n, size):
    trajectories = np.zeros((n, size), dtype=np.uint8)

    for i in range(0, n):
        # initial variables
        pi_n = pi_0.copy()
        T_i = T.copy()
        for j in range(0, size):
            # choose randomly according to the probabilities 
            c = np.random.choice(states, 1, p=pi_n)
            trajectories[i, j] = c
            
            # \pi_n = \pi_0 * T^n
            T_i = np.dot(T_i, T)
            pi_n = np.dot(pi_0, T_i)

    means = np.zeros((n,), dtype=np.float32)
    for idx, t in enumerate(trajectories):
        means[idx] = np.mean(t)

    return means


T = np.array([[0.4, 0.2, 0.1, 0, 0.1, 0.2],
             [0.2, 0.4, 0.2, 0.1, 0, 0.1],
             [0.1, 0.2, 0.4, 0.2, 0.1, 0],
             [0, 0.1, 0.2, 0.4, 0.2, 0.1],
             [0.1, 0, 0.1, 0.2, 0.4, 0.2],
             [0.2, 0.1, 0, 0.1, 0.2, 0.4]])

pi_0 = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
states = np.array([1, 2, 3, 4, 5, 6])

n = 1000
size = 30
means_30 = markov_chains(T, pi_0, states, n, size)

print(f'The empirical mean: {np.mean(means_30)}')
print(f'The empirical std: {np.std(means_30)}')
plt.hist(means_30, bins=20, density=True, alpha=0.2)
plt.title(f'Markov chains of length {size}')
plt.xlabel('Average Values')
plt.ylabel('Frequency')
plt.show()
```

#### 4.D
Write a python code to calculate $Cov(X_0,X_i)$ for $i=1,..,500$ based on $\pi_0$ and $T$. 

Compute the values for our case.

```python
def covariance(index, pi_0, states, T):
    X_mean, Y_mean, XY_mean = 0, 0, 0
    T_i = T.copy()

    count = 1
    while count < index:
        T_i = np.dot(T_i, T)
        count += 1

    pi_n = np.dot(pi_0, T_i)

    # E(X_0)
    for x, p in zip(states, pi_0):
        X_mean += x * p

    # E(X_index)
    for x, p in zip(states, pi_n):
        Y_mean += x * p

    # E(X_0*X_index)
    for x in states:
        for y in states:
            XY_mean += x * y * T_i[x - 1, y - 1]

    return XY_mean - X_mean * Y_mean


i = 1
print(covariance(i, pi_0, states, T))
```

### Question 5 - Sampling of Kendall correlation (15 points)

Recall that the Naive time complexity of computing Kendall correlation is $O(n^2)$. We will investigate a sampling approach to approximate Kendall correlations efficiently. 

#### 5.A
Draw a sample of 10M samples from a bivariate distribution $(X,Y)\sim \mathcal{N}(\bf{0},S)$ with the covariance matrix $S=\begin{pmatrix} 2 & 3\\ 3 & 5 \end{pmatrix}$. <br>
Calculate the Kendall correlation $\tau(X,Y)$.
How long did this calculation take?

```python
# visualization configuration
plt.rcParams['figure.figsize'] = (8.0, 6.0)

# initial variables
n = 10000000
means = np.array([0, 0])
S = np.array([[2, 3], [3, 5]])
sample = stats.multivariate_normal.rvs(mean=means, cov=S, size=n)

# calculating the time for activating the code
start_time = time.time()

# kendall correlation computation
X, Y = sample[:, 0], sample[:, 1]
k, _ = stats.kendalltau(X, Y)
print(f'Kendall correlation result: {k}')

end_time = time.time()
k_hours, k_rem = divmod(end_time - start_time, 3600)
k_minutes, k_seconds = divmod(k_rem, 60)
print('Calculation time: {:0>2}:{:0>2}:{:05.2f}'.format(int(k_hours), int(k_minutes), k_seconds))
```

#### 5.B
Sample 1,000 points from the data and calculate the Kendall correlation $\tau(U,V)$. 

```python
# initial the sample size variable
sample_size = 1000

# choose randomly 'sample_size' indices for X respectively to Y
idx = np.random.choice(np.arange(len(X)), size=sample_size)

# draw samples according to the indices
U = X[idx]
V = Y[idx]

# compute kendall correlation
k_1k, _ = stats.kendalltau(U, V)
print(f'Kendall correlation result: {k_1k}')
```

#### 5.C
Repeat Section 5.B 1000 times and plot a histogram of the calculated $\tau(U,V)$ values. Indicate the true $\tau(X,Y)$ (calculated on the entire 10M samples) on the histogram. What are the empirical mean and stadard deviation of the calculated $\tau(U,V)$ values?

```python
np.seterr(divide='ignore', invalid='ignore')

count = 0
samples_size, times = 1000, 1000
kendall_correlations = np.zeros((times, ), dtype=np.float16)

while count < times:
    # choose randomly 'samples_size' indices for X respectively to Y
    idx = np.random.choice(np.arange(len(X)), size=samples_size)
    
    U = X[idx]
    V = Y[idx]

    k_t, _ = stats.kendalltau(U, V)
    kendall_correlations[count] = k_t
    count += 1

print(f'The empirical mean: {np.mean(kendall_correlations)}')
print(f'The empirical std: {np.std(kendall_correlations)}')

plt.hist(kendall_correlations, bins=100, density=True, alpha=0.2)
plt.axvline(k, 0, 1, label='true kendall(X,Y) for 1M samples')
plt.title(f'Calculate {times} times kendall reuslt, for {samples_size} samples')
plt.xlabel('Kendall Results')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

#### 5.D
Repeat Section 5.C for samples of $m={10, 50, 100, 500, 1000, 5000}$ points. <br>
What is the naive time complexity of each calculation? <br>
Compare the empirical time complexity and the standard deviation of the calucated values to the size of the sample.

```python
def compute_kendall_samples(X, Y, samples_size, times):
    kendall_correlations = np.zeros((times,), dtype=np.float16)

    count = 0
    while count < times:
        idx = np.random.choice(np.arange(len(X)), size=samples_size)
        U = X[idx]
        V = Y[idx]
        k_t, _ = stats.kendalltau(U, V)
        kendall_correlations[count] = k_t
        count += 1

    return kendall_correlations


f, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(16.0, 8.0))
axes = np.array([ax1, ax2, ax3, ax4, ax5, ax6])

index = 0
times = 1000
m = np.array([10, 50, 100, 500, 1000, 5000])
time_complx = np.zeros(len(m))
stds = np.zeros(len(m))

for size, ax in zip(m, axes):
    print(f'-------- Samples size: {size} --------')

    # calculate the total samples
    print(f'Total samples(~): {size * times}', f'({((size * times) / n) * 100}% from the total data)')

    # compute kendal correlation for the given sample size
    start_time = time.time()
    kendall_correlations = compute_kendall_samples(X, Y, size, times)

    # calculate the time
    end_time = time.time()
    hours, rem = divmod(end_time - start_time, 3600)
    minutes, seconds = divmod(rem, 60)
    time_complx[index] = seconds
    print('Calculation time: {:0>2}:{:0>2}:{:05.2f}'
          .format(int(hours), int(minutes), seconds),
          f'({(seconds / k_seconds) * 100:.2f}%, relative to the true correlation time)')

    # compute mean and std
    print(f'The empirical mean: {np.mean(kendall_correlations)}')
    print(f'The empirical std: {np.std(kendall_correlations)}\n')
    stds[index] = np.std(kendall_correlations)

    # visualization
    ax.hist(kendall_correlations, bins=100, density=True, alpha=0.2), ax.set_title(f'Samples size: {size}')
    ax.axvline(k, 0, 1, label='true kendall(X,Y) for 1M samples')
    if size > 100:
        ax.set_xlabel('Kendall Results')
    ax.set_ylabel('Frequency')
    ax.legend()
    
    index += 1

plt.show()
```

```python
# plot the time complexity and the standard deviation compared to the sample size
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(16.0, 6.0))
ax1.plot(np.arange(6), time_complx)
ax1.scatter(np.arange(6), time_complx)
ax1.set_xlabel('Sample size')
ax1.set_ylabel('Time complexity (seconds)')
ax1.set_xticks(np.arange(6))
ax1.set_xticklabels([10, 50, 100, 500, 1000, 5000])

ax2.plot(np.arange(6), stds, c='g')
ax2.scatter(np.arange(6), stds, c='g')
ax2.set_xlabel('Sample size')
ax2.set_ylabel('Standard Deviation')
ax2.set_xticks(np.arange(6))
ax2.set_xticklabels([10, 50, 100, 500, 1000, 5000])

plt.show()
```

###### Based on the results
* <b>Right Chart:</b> The larger the sample size, the higher the time complexity.
* <b>Left Chart:</b> The larger the sample size, the lower the standard deviation.

###### The naive time complexity of the Kendall correlation is $O(n^2)$, however, we learned in class that we can perform the Kendall correlation algorithm with a time complexity of $O(nlog(n))$.

According to the question, "a colleague found an old dataset which she believes originated from the same population as the data in Section 5.A", we declare that our null model is that the old dataset ("Kendall_data.csv") is originated from the same population as the data in section 5.A. \
Alternatively, our alternative model states that the old dataset ("Kendall_data.csv") is <b>not</b> originated from the same population as the data in section 5.A.

$H_0 =$ The data from the "Kendall_data.csv" is originated from the same population as the data in section 5.A. \
$H_0 \neq H_A =$ The data from the "Kendall_data.csv" is <b>not</b> originated from the same population as the data in section 5.A.

### Option 1 - The Central Limit Theorem (CLT)

I will use the Central Limit Theorem (CLT), which states that the distribution of sample approximates a standard normal distribution, as the sample size becomes larger.

The mean of a sample of data will be closer to the mean of the overall population in question, as the sample size increases, notwithstanding the actual distribution of the data. \
Therefore, I used the empirical mean and the empirical standard deviation from the Kendall results (in section 5.D), to compute the p-value under the null model, using the CDF of a standard normal distribution. 

<b>The Central Limit Theorem:</b> 
$\lim_{n \to \infty} P\left(\frac{\bar{X}_n-\mu}{\sigma/\sqrt{n}} \leq z\right) = \Phi(z)$

<b>Notations</b> \
$\bar{x}=\frac{1}{N}\sum_{i=1}^{N}x_{i}$ \
$\hat{\sigma}^2 = \frac{1}{N-1} \sum_{i=1}^N \left(x_i - \bar{x}\right)^2$ 

<b>Calculations</b> \
$\mu = \bar{x}_n$ \
$\sigma = \hat{\sigma}$ \
$\lim_{n \to \infty} P\left(\frac{\bar{X}_n-\mu}{\sigma/\sqrt{n}} \leq \frac{\tau_{colleague} - \mu}{\sigma} \right) = \Phi(\frac{\tau_{colleague} - \mu}{\sigma})$

### Results
According to the results below, we <b>cannot reject</b> our null model since our p-value is 0.2839 (which is not a significant statistic value).

### Option 2 - Confidence Interval

I will use the Confidence Interval computation to calculate the interval for the empirical mean in our data from section 5.D. Once this is done, I will have the interval for the empirical mean that define on which interval is the real mean exists, and to reject our null model, I'll need to find the correct confidence which the Kendall correlation (from the "Kendall_data.csv") is not within the interval. 

According to the results below, we <b>cannot reject</b> our null model since the result of the confidence interval is 42% (lower confidence). \
Respectively, it means there is a confidence of 58% that the Kendall correlation result from the "Kendall_data.csv" data is originated from our data in section 5.A).

```python
# computes the mean and the std of the kendall correlations data
p_rohat_k = np.mean(kendall_correlations)
sigma_rohat_k = kendall_correlations.std(ddof=1)

print(f'The empirical mean of the kendall correlations: {p_rohat_k:.4f}')

# define the alpha for the confidence interval
alpha = 0.58  # 1- 0.58 = 0.42
gamma = stats.norm.ppf(q=1 - alpha / 2)

# calculate the confidence interval for the empirical mean
error_margin = gamma * sigma_rohat_k
lower_bound = p_rohat_k - error_margin
upper_bound = p_rohat_k + error_margin
kendall_interval = (lower_bound, upper_bound)
print(f'The mean of the kendall correlations is in the interval: {kendall_interval}')

# the old data according to the null hypothesis
df_kendall_data = pd.read_csv('Kendall_data.csv', names=['X', 'Y'])
X2, Y2 = df_kendall_data['X'], df_kendall_data['Y']
k2, _ = stats.kendalltau(X2, Y2)

print(f'The kendal correlation result for Kendall_data.csv is: {k2}')
print(f'The kendall correlation result: {k2} (based on the Kendall_data.csv file) is not inside the interval' if not (lower_bound < k2 < upper_bound) else '')
```
